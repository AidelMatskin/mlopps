```python
make_dataframe = '''SELECT * FROM housing_data'''
df = pd.read_sql(make_dataframe, conn)
```

```python
# changing all yes/no to 1/0 so that it can be used in model
mapping = {'yes': 1, 'no': 0}
columns_to_map = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[columns_to_map]= df[columns_to_map].replace(mapping)
```

```python
#making dummy columns for furnished so that it can be used in model

df= pd.get_dummies(df, columns =['furnishingstatus'])
```

```python
df.head()
```

```python
X=df.drop('price',axis=1)
y= df['price']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
lr.score(X_test, y_test)
```

```python
MAE= mean_absolute_error(y_test, predictions)
print(MAE)
```

```python
MSE = mean_squared_error(y_test, predictions)
print(MSE)
```

```python
RMSE = mean_squared_error(y_test, predictions, squared = False)
print(RMSE)
```
